```python
def get_dialogue_label_sequence(dialogue):
    dialog_label_sequence = []
    for i,j in dialogue.iterrows():
        if isinstance(j.er_label_1, str):
            dialog_label_sequence.append(j.er_label_1 + '_er')
        else:
            dialog_label_sequence.append(j.ee_label_1 + '_ee')
    return dialog_label_sequence

dialogue_label_sequence = get_dialogue_label_sequence(dialogue)
```

```python
def ngrams(sequence, n=2):
    output = {}
    for i in range(len(sequence)-n+1):
        g = ' '.join(sequence[i:i+n])
        output.setdefault(g, 0)
        output[g] += 1
    return output

def next_label_probability(previous, sequence, n=2):
    prob = {}
    prev_count = ngrams(sequence, 1)[previous]
    grams = ngrams(sequence, n)

    for gram in grams:
        if gram[ 0 : len(previous) ] == previous:
            prob[gram[len(previous)+1:]] = (grams[gram]/prev_count)
    
    return prob 
```

```python
# Sort intent and write to .md file with correct format for rasa NLU training

intents_for_classification = []

for i,j in annotated_persuasion.iterrows():
    if isinstance(j.er_label_1, str):
        intents_for_classification.append([j.Unit, j.er_label_1])

intents_for_classification.sort(key=lambda x : x[1])
    
```

```python
f = open("demofile1.txt", "w")

for i in range(0, len(intents_for_classification)):
    if i == 0: 
        f.write(str("## intent:" + intents_for_classification[i][1] + "\n"))
        
    elif intents_for_classification[i][1] != intents_for_classification[i-1][1]:
        f.write(str("\n ## intent:" + intents_for_classification[i][1] + "\n"))
    
    f.write(str("- " + intents_for_classification[i][0] + "\n"))
    
        
f.close()
```

```python
def get_dialog_as_string(dialogue_labels):
    instring = ''
    for turn in dialogue_labels:
        instring = instring + ' ' + turn
    return instring
```

```python
conversationID = annotated_persuasion.B2.unique()
metadata = pd.read_excel('300_info.xlsx')

convos = []
for conv in conversationID:
    a = annotated_persuasion[annotated_persuasion['B2'] == conv ]
    donation = metadata[metadata['B2'] == conv ].B6
    dialog_labels = get_dialogue_label_sequence(a)
    dialog_as_string = get_dialog_as_string(dialog_labels)
    
    convos.append([conv, dialog_as_string, len(annotated_persuasion[annotated_persuasion['B2'] == conv]), donation])
    
data = pd.DataFrame(convos, columns = ['ID', 'sequence', 'length', 'donation'])
```
